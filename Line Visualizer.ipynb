```python
import cv2
import numpy as np
import math

image1 = cv2.imread('/home/mohit.srivastava@Brightlifecare.local/Downloads/t.png')
gray=cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)
canimg = cv2.Canny(gray, 50, 200)

lines= cv2.HoughLines(canimg, 1, np.pi/180.0, 120, np.array([]))


for line in lines:
    rho, theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))

    cv2.line(image1,(x1,y1),(x2,y2),(0,0,255),2)


#cv2.imshow('Lines Detected',image1)
cv2.imshow("Canny Detection", canimg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the test tube image
test_tube_image = cv2.imread("/home/mohit.srivastava@Brightlifecare.local/Downloads/t.png")

# Convert the image to grayscale
gray_image = cv2.cvtColor(test_tube_image, cv2.COLOR_BGR2GRAY)

# Apply adaptive thresholding to segment the black line markings
thresholded_image = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 4)

# Find contours in the thresholded image
contours, _ = cv2.findContours(thresholded_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Iterate over the contours and filter out the ones that match line-like shapes
line_contours = []
for contour in contours:
    # Approximate the contour to a polygon
    epsilon = 0.01 * cv2.arcLength(contour, True)
    approx = cv2.approxPolyDP(contour, epsilon, True)
    
    # Filter contours based on number of vertices (assuming lines have 2 vertices)
    if len(approx) == 2:
        line_contours.append(approx)

# Draw the detected line contours on the original image
detected_lines_image = cv2.drawContours(test_tube_image, line_contours, -1, (0, 255, 0), 2)

# Display the results
cv2.imshow("Original Image", test_tube_image)
cv2.imshow("Thresholded Image", thresholded_image)
cv2.imshow("Detected Lines", detected_lines_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the video
video_path = "/home/mohit.srivastava@Brightlifecare.local/Downloads/v.mp4"
cap = cv2.VideoCapture(video_path)

# Check if the video is successfully loaded
if not cap.isOpened():
    print("Error opening video file.")
    exit()

# Define the codec and create a VideoWriter object to save the output video
codec = cv2.VideoWriter_fourcc(*"mp4v")
output_video = cv2.VideoWriter("output_video.mp4", codec, 30, (640, 480))

# Process each frame in the video
while cap.isOpened():
    # Read the current frame
    ret, frame = cap.read()

    # If the frame is not read properly, break the loop
    if not ret:
        break

    # Convert the frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Apply adaptive thresholding to segment the black line markings
    thresholded_frame = cv2.adaptiveThreshold(gray_frame, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 4)

    # Find contours in the thresholded frame
    contours, _ = cv2.findContours(thresholded_frame, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Iterate over the contours and filter out the ones that match line-like shapes
    line_contours = []
    for contour in contours:
        # Approximate the contour to a polygon
        epsilon = 0.01 * cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, epsilon, True)

        # Filter contours based on number of vertices (assuming lines have 2 vertices)
        if len(approx) == 2:
            line_contours.append(approx)

    # Draw the detected line contours on the frame
    detected_lines_frame = cv2.drawContours(frame, line_contours, -1, (0, 255, 0), 2)

    # Display the resulting frame
    cv2.imshow("Detected Lines", detected_lines_frame)

    # Write the frame to the output video
    output_video.write(detected_lines_frame)

    # Break the loop if the 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the resources
cap.release()
output_video.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the video
video_path = "/home/mohit.srivastava@Brightlifecare.local/Downloads/v.mp4"
cap = cv2.VideoCapture(video_path)

# Check if the video is successfully loaded
if not cap.isOpened():
    print("Error opening video file.")
    exit()

# Define the codec and create a VideoWriter object to save the output video
codec = cv2.VideoWriter_fourcc(*"mp4v")
output_video = cv2.VideoWriter("output_video.mp4", codec, 30, (640, 480))

# Process each frame in the video
while cap.isOpened():
    # Read the current frame
    ret, frame = cap.read()

    # If the frame is not read properly, break the loop
    if not ret:
        break

    # Convert the frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Apply adaptive thresholding to segment the black line markings
    thresholded_frame = cv2.adaptiveThreshold(gray_frame, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 11)

    # Find contours in the thresholded frame
    contours, _ = cv2.findContours(thresholded_frame, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Iterate over the contours and filter out the horizontal line contours
    line_contours = []
    for contour in contours:
        # Compute the bounding box for the contour
        x, y, w, h = cv2.boundingRect(contour)

        # Filter contours based on aspect ratio and width threshold
        aspect_ratio = w / float(h)
        if aspect_ratio > 5 and h > 10:
            line_contours.append(contour)

    # Draw the detected line contours on the frame
    detected_lines_frame = cv2.drawContours(frame, line_contours, -1, (0, 255, 0), 25)

    # Display the resulting frame
    cv2.imshow("Detected Lines", detected_lines_frame)

    # Write the frame to the output video
    output_video.write(detected_lines_frame)

    # Break the loop if the 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the resources
cap.release()
output_video.release()
cv2.destroyAllWindows()
```
