```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('/home/mohit.srivastava@Brightlifecare.local/test_images/67c.png')
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)

edges = cv.Canny(gray, 80, 220, apertureSize = 3)
plt.imshow(edges, cmap='gray')
```

```python
lines = cv.HoughLines(edges, 1, np.pi/180, 50)
print(len(lines))
```

```python
lines = cv.HoughLinesP(image=edges,rho=1,theta=np.pi/180, threshold=30, 
                       lines=np.array([]), minLineLength=10, maxLineGap=10)
print(len(lines))
```
